```python
# import statements for all of the packages that i plan to use.
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
read_file = pd.read_csv('noshowappointments-kagglev2-may-2016.csv',sep=';')
read_file.head(10)
```

```python
ENGletters = ['F']
read_file = read_file[~read_file['Gender'].str.contains('|'.join(ENGletters))]
read_file = read_file.sort_index()
read_file = read_file.reset_index(drop=True)
read_file
```

```python
data = []
for i in range(len(read_file)):
        id = read_file['Neighbourhood'][i]
        text = read_file['Age'][i]            
        data.append({"audio": id, "text": text})
```
